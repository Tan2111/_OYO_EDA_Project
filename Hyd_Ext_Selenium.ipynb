```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import requests
from bs4 import BeautifulSoup

import re
import time

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
Name_H = [] 
Cost = []
Rate = []
Feature = []
Add = []
# go throught the 9 pages
for i in range(1,10):
        # luanch a chrome instance
    link = 'https://www.oyorooms.com/hotels-in-hyderabad/?page={}/'.format(i)
    driver.get(link)
    driver.maximize_window() # maximize the window
    sleep(2) # wait for the page to load
    
      # Load the HTML of the page
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    

    for k in soup.find_all('div',class_="oyo-cell--12-col listingHotelDescription__priceBtn"):
        if k is not None:
            Cost.append(k.text)
        else:
            Cost.append(np.NaN)
            
            
    for a1 in soup.find_all('h3', attrs={'class':'listingHotelDescription__hotelName d-textEllipsis'}):
        if a1 is not None:
            Name_H.append(a1.text)
        else:
            Name_H.append(np.NaN)
    
    for c1 in soup.find_all('div', attrs={'class':'hotelRating'}):
        if c1 is not None:
            Rate.append(c1.text)
        else:
            Rate.append(np.NaN)
    
    for d1 in soup.find_all('div', attrs={'class':'amenityWrapper'}):
        if d1 is not None:
            Feature.append(d1.text)
        else:
            Feature.append(np.NaN)
        
    for E1 in soup.find_all('div', attrs={'d-body-lg listingHotelDescription__hotelAddress'}):
        if E1 is not None:
            Add.append(E1.text)
        else:
            Add.append(np.NaN)
    
```

```python
print(len(Name_H[:148]))
print(len(Cost[:148]))
print(len(Rate[:148]))
print(len(Feature))
print(len(Add[:148]))
```

```python
df = pd.DataFrame({'Hotel Name' : Name_H[:148], 'Cost Per Night' : Cost[:148], 'Rating' : Rate[:148], 'Features' : Feature, 'Location': Add[:148]})
```
